## web scraping 

```python
import requests as req # to get web page  

from bs4 import BeautifulSoup # to parsing html code or page web (extract data from web page)

import csv

import pandas as pd
```

```python
# get a page 

page=req.get("https://www.yallakora.com/match-center/?date=11/22/2022#days")
```

```python
def main (page) :
    
    # to get page content
    con=page.content
    
    # to convert from byte code into html code 
    soup=BeautifulSoup(con,"lxml")
    
    #create list to what i need from this data
    matches_detail=[]
    
    #import data from (div class) that contain data that i need (div class( #tag) exist in section)
    
    # tag or div class has a descendants or (childern) these childern has the data that i need 
   
    
    championships=soup.find_all('div',{'class':'matchCard'})
    #championship is list has all div class 
    
    #contents is list has a childern or tags that exist in div class and we can dont use it 
    
    
    for ii in range (len(championships)):   
        # find name_of_champions 
        championship_title=championships[ii].contents[1].find('h2').text.strip()
    
    
        # find matches    
        matches=championships[ii].contents[3].find_all('li')
   
        number_of_matches=len(matches)
    
        for i in range(number_of_matches) :
        
            # find the first team
            team_A=matches[i].find('div',{'class':'teamA'}).text.strip()
        
            #find the second team
            team_B=matches[i].find('div',{'class':'teamB'}).text.strip()
        
            #find the result or score
            result=matches[i].find('div',{'class':'MResult'}).find_all('span',{'class':'score'})[0].text.strip()+"_"+ matches[i].find('div',{'class':'MResult'}).find_all('span',{'class':'score'})[1].text.strip() 
           
            #find time of match
            Time=matches[i].find('div',{'class':'MResult'}).find('span',{'class':'time'}).text.strip()
        
            #Add matche info to list  
            matches_detail.append({'نوع البطولة':championship_title,'الفريق الاول':team_A,'النتيحة':result,'الفريق الثاني':team_B,'وقت المبارة':Time})
   
    keys1=matches_detail[0].keys()  
    with open('C:\\Users\\HP\\Desktop\\web scraping\\world-cup.csv','w',encoding='utf-8-sig') as matches_csv :
        dict_writer=csv.DictWriter(matches_csv,keys1)
        dict_writer.writeheader()
        dict_writer.writerows(matches_detail)
```

```python

    
main(page)
```
