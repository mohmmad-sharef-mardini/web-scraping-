```python
import requests as req 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
page=req.get('https://en.wikipedia.org/wiki/List_of_largest_companies_of_the_United_Arab_Emirates')
con=page.content
soup=BeautifulSoup(con,'lxml')
```

```python
def header(s):
    table=s.find_all('table',{'class':'wikitable'})[1]
    columns=table.find('tr').find_all('th')
    column=[]
    for i in range(len(columns)) :
        column.append(columns[i].text.strip())
    
    return column
    
    
    

     
    
```

```python
df=pd.DataFrame(columns=header(soup))
```

```python
def rows (s):
    table=s.find_all('table',{'class':'wikitable'})[1]
   
    
    rows=table.find('tbody').find_all('tr')
    
    
    for i in range(len(rows)) :
        
        if i == 0:
            continue 
        
        
        re=rows[i].find_all('td')
        
        row=[]
        for ii in range(len(re)) :
            filed=re[ii]
            row.append(filed.text.strip())
        
        
    
        length=len(df)
        df.loc[length]=row




rows(soup)    
```
